```julia
#Include the UNSflow libraries
#Pkg.clone("git://github.com/KiranUofG/UNSflow")
include("../src/UNSflow.jl")
using UNSflow
```

### Define all the kinematics.





```julia
fstar = 0.14 #nondimensional flapping frequency fstar = fc/U
theta0 = 76.3 #pitch amplitude in deg
h0 = 1.0 #plunge amplitude

#Sinusoidal pitch:
    #mean = 0
    #amplitude = theta0 
    #k = fstar
    #phi = 0 deg
    
alphadef = SinDef(0.,theta0*pi/180,fstar*pi,0.)

#Sinusoidal plunge:
    #mean = 0
    #amplitude = theta0
    #k = fstar
    #phi = 90 deg
    
hdef = SinDef(0.,h0,fstar*pi,90*pi/180)

#Freestream velocity is a constant. Set as constant = 1
udef = ConstDef(1.)

#The full kinematics are collected as:
full_kinem = KinemDef(alphadef, hdef, udef)
```

### Define the surface undergoing motion and the flowfield. The LESP_crit used below is the critical LESP parameter which is used to predict and modulate LEV shedding - it depends on the airfooil shape and Reynolds number of operation. See paper for more details.

```julia
pvt = 0.333 #

lespcrit = [0.19;] #Note that the value is entered as an array. 

# The argument "FlatPlate" provides zero camber. To use an airfoil, specify the coordinate file which should be
#in the same folder. For eg. "sd7003.dat". 
#The XFOIL convention must be used for the coordinate: TE->Upper surface->LE->Lower surface->TE
surf = TwoDSurf("naca0015.dat", pvt, full_kinem,lespcrit)

curfield = TwoDFlowField()
```

### Specify simulation details and run code. 

```julia
#Specify simulation details - time step, and number of steps.
#The default nondimensional timestep is 0.015. There is usually no need to change this. 
ncyc = 2 #Number of cycles to be simulated
time_total = 1/0.14*ncyc #Required total time
nsteps = Int(round(time_total/0.015))
```

```julia
#Now the code is run as :
#The variables mat, surf and curfield are returned
mat, surf, curfield = ldvm(surf, curfield, nsteps)
```

```julia
alpha1 = 15.25*pi/180
fsep_noabs, x_duh_noabs, x_duh, h_duh, mat_alt, surf_alt, curfield_alt = ldvm_klb_alt(surf, curfield, nsteps, 0.015; alpha1)
```

```julia
mat_klb, surf_klb, curfield_klb = ldvm_klb(surf, curfield, nsteps)
```

### Studying the results

```julia
range_full = round(Int,(ncyc-1)*nsteps/ncyc)+1:nsteps
tbyT=linspace(0,1,length(range_full))

plot(tbyT,mat[range_full,6], label="No TEFSC")
plot(tbyT,mat_klb[range_full,6], label="TEFSC old")
#plot(mat_alt[:,1],mat_alt[:,6], label="TEFSC new with abs")

xlabel("Nondimensional time t/T")
ylabel("Lift coefficient CL")
legend()
```
