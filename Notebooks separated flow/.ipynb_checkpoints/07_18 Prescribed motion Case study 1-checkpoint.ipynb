#### This notebook tests LDVM solver with KLB for prescribed motions. Results are compared to the case study 1 from [1]. Parameters of motion are presented in Table 1.

[1] K. Ramesh, A. Gopalarathnam, K. Granlund, M. V. Ol, and J. R. Edwards, “Discrete-vortex method with novel shedding criterion for unsteady aerofoil flows with intermittent leading-edge vortex shedding,” Journal of Fluid Mechanics, vol. 751, pp. 500–538, Jul. 2014.


#### Table 1
|--|--|
|:----:|:-------:|
|Airfoil|SD-7003|
|Motion|Eldredge ramp|
|Smoothing parameter a|11|
|Amplitude|25 deg|
|Pivot point| LE|
|Nondim. pitch rate k|0.11|
|Reynolds number| 30k|
|LESP| 0.18|

```julia
# Insert stating constants

    #Static constants from case 6 (notebook "KLB - static SD7003 test cases") are used
    alpha1, s1, s2, model = 10.5,        3.06665,  3.70008, "Original"
    param = [-0.0704412  -0.011237  0.0739017  1.88094]
        # 1: k0, 2: k1, 3: k2, 4: m
    #Parameters from [2] for NACA 0015: alpha1 = 15.25, s1 = 3.0, s2 = 2.3, k0 = 0, k1 = -0.135, k2 = 0.04, m = 2 
    #[2] Z. Liu, J. C. S. Lai, J. Young, and F.-B. Tian, “Discrete Vortex Method with Flow Separation Corrections for Flapping-Foil Power Generators,” AIAA Journal, vol. 55, no. 2, pp. 410–418, Feb. 2017.
```

```julia
alphadef = EldRampReturnDef(25.*pi/180.,0.11, 11.)
hdef = ConstDef(0.)
udef = ConstDef(1.) #Freestream velocity is a constant. Set as constant = 1

#The full kinematics are collected as:
full_kinem = KinemDef(alphadef, hdef, udef)

pvt = 0. #Leading edge
lespcrit = [0.18;] #Note that the value is entered as an array. 

# The argument "FlatPlate" provides zero camber. To use an airfoil, specify the coordinate file which should be
#in the same folder. For eg. "sd7003.dat". 
#The XFOIL convention must be used for the coordinate: TE->Upper surface->LE->Lower surface->TE
surf = TwoDSurf("sd7003_fine.dat", pvt, full_kinem,lespcrit)
curfield = TwoDFlowField()
del = DelVortDef(1, 500, 10)

#Specify simulation details - time step, and number of steps.
#The default nondimensional timestep is 0.015. There is usually no need to change this. 
dtstar = find_tstep(alphadef)
time_total = 7.
nsteps = Int(round(time_total/dtstar))

#Now the code is run as :
#The variables mat, surf and curfield are returned
        
mat1d, surf, curfield = ldvm_klb_tr(surf, curfield,param[1,:], nsteps, dtstar; delvort = del, alpha1 = alpha1, S1 = s1, S2 = s2, model=model, aoa = 1, modelF = "dymore");
```
