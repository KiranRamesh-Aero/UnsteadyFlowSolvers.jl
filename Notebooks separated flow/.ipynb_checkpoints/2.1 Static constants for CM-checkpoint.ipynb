### This notebook shows how k0, k1, k2 and m static constants required for moment coefficient calculation are obtained. 

Input: CM($\alpha$), CN($\alpha$) characteristics from Xfoil; f($\alpha$)

Output: $k_0$, $k_1$, $k_2$, m

```julia
#Load data from Xfoil
xfoil = Xfoil("sd7003_data.txt");

#Calculate f basing on CN(alpha) data from Xfoil
f = fFromCN(xfoil.alpha, xfoil.CN);
```

```julia
#Obtain static constants
static_constants = cmstatic(xfoil.CM, xfoil.CN, f; par = 1.0)
```

### Procedure implemented in cmstatic function is now explained step by step:

Fitting function is defined as:
 
Where k0 = (0.25 - xac) is aerodynamic center offset from the quarter chord, k1 represents the direct effect on the center of pressure due to the growth of the separated flow region, and k2 helps describe the shape of the moment break at stall. The values of k0; k1; k2 and m can be adjusted to give the best moment reconstruction.

Source: O. A. Bauchau, “DYMORE user’s manual,” Georgia Institute of Technology, Atlanta, 2007.

```julia
fun(f, p) = p[1] + p[2]*(1.-f)+p[3]*sin(pi.*f.^p[4]) #1 - k0, 2 - k1, 3 - k2, 4 - m
init = [0., -0.135, 0.04, 2.] #initial values from [1] Z. Liu, J. C. S. Lai, J. Young, and F.-B. Tian, 
                            #“Discrete Vortex Method with Flow Separation Corrections for Flapping-Foil Power Generators,” 
                            #AIAA Journal, vol. 55, no. 2, pp. 410–418, Feb. 2017.
fit = curve_fit(fun, f, xfoil.CM./xfoil.CN, init)
fit.param
```

```julia
plot(f,xfoil.CM./xfoil.CN)
plot(f,fun(f,fit.param))
xlabel("Separation point f as x/c")
ylabel("CM/CN")
annot = string("k0 = ", string(fit.param[1])[1:5],"\n","k1 = ", string(fit.param[2])[1:5],"\n","k2 = ", string(fit.param[3])[1:5],"\n","m = ", string(fit.param[4])[1:3])
            annotate(annot,
            xy=[1;0],
            xycoords="axes fraction",
            xytext=[-220,10],
            textcoords="offset points",
            fontsize=15.0,
            ha="right",
            va="bottom")
```

#### It can be noticed that better fit may be obtained by limiting the domain, in this case down to f = (0 , 0.9). Upper limit of the domain is the keyword parameter "param" of the function cmstatic.

```julia
idx = 1
while(f[idx]>0.9)
    idx+=1
end

fit = curve_fit(fun, f[idx:end], xfoil.CM[idx:end]./xfoil.CN[idx:end], init)
fit.param

plot(f,xfoil.CM./xfoil.CN)
plot(f,fun(f,fit.param))
xlabel("Separation point f as x/c")
ylabel("CM/CN")
annot = string("k0 = ", string(fit.param[1])[1:5],"\n","k1 = ", string(fit.param[2])[1:5],"\n","k2 = ", string(fit.param[3])[1:5],"\n","m = ", string(fit.param[4])[1:3])
            annotate(annot,
            xy=[1;0],
            xycoords="axes fraction",
            xytext=[-220,10],
            textcoords="offset points",
            fontsize=15.0,
            ha="right",
            va="bottom")
```
