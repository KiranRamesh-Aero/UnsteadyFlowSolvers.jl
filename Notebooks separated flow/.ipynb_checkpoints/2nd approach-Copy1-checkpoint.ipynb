```julia
#only 00XX NACA supported at the moment
naca = "0012"
c= 1

max_thickness = parse(Int,naca)

#Preview airfoil
f(x) = 5.*max_thickness./100.*(0.2969.*sqrt(x).-0.1260.*x-0.3516.*x.^2+0.2843.*x.^3-0.1015.*x.^4)

x = linspace(0, 1, 250)
plot(x,f(x))
axis("equal")
```

```julia
x = Sym("x")
a0 = 0.2969
a1 = -0.1260
a2 = -0.3516
a3 = 0.2843
a4 = -0.1015
```

```julia
uTbyU(x) = 5.*max_thickness./100.*((a1.+a2.+a3.*0.75.+a4.*0.5)./(2.*pi).*log((1.+x)./(1.-x)).+(a2.+a3.*1.5.+a4.*1.5)./(2.*pi).*(x.*log((1.+x)./(1.-x)).-2).+(0.75.*a3.+1.5.*a4)./(2.*pi).*(x.*(x.*log((1.+x)./(1.-x)).-2)).+(a4./(4.*pi)).*(x.^2.*(x.*log((1.+x)./(1.-x)).-2).-2./3).+(a0./(2.*sqrt(2).*pi)).*(1./sqrt(1.+x)).*log((sqrt(2).+sqrt(1.+x))./(sqrt(2).-sqrt(1.+x))))
```

```julia
tau = Sym("tau")
b1 = Sym("b1")
b4 = Sym("b4")
b6 = Sym("b6")
b8 = Sym("b8")
```

```julia
q1t(x,tau,b1,b4,b6,b8) = tau./pi.*(1./tau.*dT(x).*log(x./(1.-x))+b1./sqrt(x).*log((1.+sqrt(x))./x).-2.*b4.-3./2.*b6.-4./3.*b8.-(3.*b6.+2.*b8).*x.-4.*b8.*x.^2)
q1t(x,tau,b1,b4,b6,b8)
```
