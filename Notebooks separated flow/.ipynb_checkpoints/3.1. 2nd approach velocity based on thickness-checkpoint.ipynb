```julia
using PyPlot, SymPy
```

```julia
#USER INPUT
naca = "0012" #only 00XX NACA supported at the moment
c= 1
```

```julia
tau = parse(Int,naca)/100
b1 = 1.48450
b2 = -0.63
b4 = -1.758
b6 = 1.4215
b8 = -0.5075

#Preview airfoil
f(x) = tau.*(b1.*sqrt(x).+b2.*x+b4.*x.^2+b6.*x.^3+b8.*x.^4)

x = linspace(0, 1, 250)
plot(x,f(x))
plot(x,-f(x))
axis("equal")
title("NACA $(naca)")
```

```julia
x=Sym("x")
#Surface speed due to thickness
dT(x) = tau.*(b1./(2.*sqrt(x)).+b2.+2.*b4.*x.+3.*b6.*x.^2.+4.*b8.*x.^3)
q1t(x) = tau./pi.*(1./tau.*dT(x).*log(x./(1.-x))+b1./sqrt(x).*log((1.+sqrt(x))./sqrt(x)).-2.*b4.-3./2.*b6.-4./3.*b8.-(3.*b6.+2.*b8).*x.-4.*b8.*x.^2)

#Result for NACA0012 from Dyke, for validation
gs(x) = -dT(x)./pi.*log((1.-x)./x)+0.17814./(pi.*sqrt(x)).*log((1.+sqrt(x))./sqrt(x)).+0.0787021.-0.12412118.*x.+0.0775403.*x.^2
```

```julia
gs(x)
```

```julia
q1t(x)
```

```julia
ro = (1.1019*tau^2*c)/2
```

```julia
coseta(x) = sqrt(x./(x.+ro./2))
plot(xnaca,coseta(xnaca))
```

```julia
xnaca = linspace(0,1,250)
plot(xnaca,1.+q1t(xnaca),label="Formal first-order theory")
#plot(xnaca,1.+gs(xnaca),label="Validation data")
plot(xnaca,coseta(xnaca).*(1.+q1t(xnaca)), label ="Formal first-order theory with Riegel's rule")
ylim(0.9,1.3)
xlim(0.,1.)
legend()
title("Speed on NACA $(naca) at zero angle of attack, incompressible")
```
