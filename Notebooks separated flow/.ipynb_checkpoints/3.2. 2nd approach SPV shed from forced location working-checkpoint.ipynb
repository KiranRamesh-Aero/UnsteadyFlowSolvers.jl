```julia
#Include the UNSflow libraries
#Pkg.clone("git://github.com/KiranUofG/UNSflow")
include("../src/UNSflow.jl")
using UNSflow
```

This notebook tests 2nd approach. SPV are forced to be shed from one location on an aerofoil that is at steady alpha = 12 deg angle of attack.

First, calculate LDVM to see flowfield with LEV

```julia
# Insert stating constants



alphadef = ConstDef(12.*pi/180)
hdef = ConstDef(0.)
udef = ConstDef(1.) #Freestream velocity is a constant. Set as constant = 1

#The full kinematics are collected as:
full_kinem = KinemDef(alphadef, hdef, udef)

pvt = 0.25 #Leading edge
lespcrit = [0.08;] #Note that the value is entered as an array. 
time_total = 3.
#nsteps = Int(round(time_total/0.015))

# The argument "FlatPlate" provides zero camber. To use an airfoil, specify the coordinate file which should be
#in the same folder. For eg. "sd7003.dat". 
#The XFOIL convention must be used for the coordinate: TE->Upper surface->LE->Lower surface->TE
nsteps=100
surf = TwoDSurf("n0012.dat", pvt, full_kinem,lespcrit)
curfield = TwoDFlowField()
#Now the code is run as :
#The variables mat, surf and curfield are returned
        
matldvm, surf, curfield = ldvm(surf, curfield, nsteps);
```

```julia
view_vorts(surf,curfield)
axis("equal")
```

```julia
plot(matldvm[:,9], label="TEV")
plot(matldvm[:,10], label="LEV")
legend()
```

Now, let's try LDVM_SPV solver with SPV vortices forced to be shed from f = 0.05

```julia
# The argument "FlatPlate" provides zero camber. To use an airfoil, specify the coordinate file which should be
#in the same folder. For eg. "sd7003.dat". 
#The XFOIL convention must be used for the coordinate: TE->Upper surface->LE->Lower surface->TE
surf = TwoDSurf("n0012.dat", pvt, full_kinem,lespcrit)
curfield = TwoDFlowField()

#Now the code is run as :
#The variables mat, surf and curfield are returned
sepdef = SeparationParams(7., 5.,5.,"Sheng", fsep_forced = 0.05);        
mat2d, surf, curfield = ldvm_spv(surf, curfield,sepdef, nsteps);
```

```julia
view_vorts(surf,curfield)
axis("equal")
```

```julia
plot(mat2d[:,11],label="TEV")
plot(mat2d[:,12],label="SPV")
legend()
```

Strengths of vortices are more or less ok in comparison to LDVM with LEVs

```julia
plot(mat2d[:,1],mat2d[:,10],label="Dynamic f")

xlabel("Nondimensional time t*")
ylabel("Separation point x/c")
legend()
```

Dynamic separation point location is constantly at f=0.05, which is correct. Now let's try the same thing, but with f = 0.5

```julia
# The argument "FlatPlate" provides zero camber. To use an airfoil, specify the coordinate file which should be
#in the same folder. For eg. "sd7003.dat". 
#The XFOIL convention must be used for the coordinate: TE->Upper surface->LE->Lower surface->TE
surf = TwoDSurf("n0012.dat", pvt, full_kinem,lespcrit)
curfield = TwoDFlowField()

#Now the code is run as :
#The variables mat, surf and curfield are returned
        
sepdef = SeparationParams(7., 5.,5.,"Sheng", fsep_forced = 0.5);        
mat2d, surf, curfield = ldvm_spv(surf, curfield,sepdef, nsteps);
```

```julia
view_vorts(surf,curfield)
axis("equal")
```

```julia
plot(mat2d[:,1],mat2d[:,10],label="Dynamic f")

#plot(mat2d[:,1],f0sep2, label="preview of static f")
xlabel("Nondimensional time t*")
ylabel("Separation point x/c")
legend()
```

It's hard to tell whether it's 100% correct, but the solution converges and the strength of vortices is reasonable. Now let's try f = 1.0.
