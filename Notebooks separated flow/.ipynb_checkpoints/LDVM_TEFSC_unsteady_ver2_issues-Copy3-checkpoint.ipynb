# DVM with TEFSC

### This study tests various implementations of effective separation point prediction for unsteady flow conditions. Test case is a flat plate undergoing prescribed sinusoidal motion in pitch and plunge.

###### DVM model used: Ramesh, K., Gopalarathnam, A., Granlund, K., Ol, M.V., Edwards, J.R., 2014. Discrete vortex method with novel shedding criterion for unsteady airfoil flows with intermittent leading-edge vortex shedding. Journal of Fluid Mechanics 751, 500–538.

###### Model used for unsteady TEFSC: Goman, M., Khabrov, A., State-Space Representation of Aerodynamic Characteristics of an Aircraft at High Angles of Attack. Journal of Aircraft, Vol. 31, No. 5, 1994, pp. 1109-1115.
###### Fan, Y. Lutze F.H., Identification of an Unsteady Aerodynamic Model at High Angles of Attack. 21st Atmospheric Flight Mechanics Conference, AIAA Paper 1996-3407, July 1996.

```julia
#Include the UNSflow libraries
#Pkg.clone("git://github.com/KiranUofG/UNSflow")
include("../src/UNSflow.jl")
using UNSflow

fstar = 0.18 #nondimensional flapping frequency fstar = fc/U
theta0 = 60. #pitch amplitude in deg
h0 = 1.

#Sinusoidal pitch:
    #mean = 0
    #amplitude = theta0 
    #k = fstar
    #phi = 0 deg
    
alphadef = SinDef(0.,theta0*pi/180,fstar*pi,0.)

#Sinusoidal plunge:
    #mean = 0
    #amplitude = theta0
    #k = fstar
    #phi = 90 deg
    
hdef = SinDef(0.,theta0*pi/180,fstar*pi,90*pi/180)

#Freestream velocity is a constant. Set as constant = 1
udef = ConstDef(1.)

#The full kinematics are collected as:
full_kinem = KinemDef(alphadef, hdef, udef)

pvt = 0.333 

lespcrit = [5;] #LEV shedding turned off 

surf = TwoDSurf("FlatPlate", pvt, full_kinem,lespcrit)

curfield = TwoDFlowField()

#Specify simulation details - time step, and number of steps.
#The default nondimensional timestep is 0.015. There is usually no need to change this. 
ncyc = 2 #Number of cycles to be simulated
time_total = 1/fstar*ncyc #Required total time
nsteps = Int(round(time_total/0.015))
```

```julia

#Now the code is run as :
#The variables mat, surf and curfield are returned
mat,surf,curfield = ldvm_klb_tr(surf, curfield, nsteps)
```

```julia
t0 = 0 
tf = 12
N = 5000
dt = (tf-t0)/N
t = t0:dt:tf
x = zeros(length(t))

h = zeros(length(t))

for k = 1 : length(t)
    if t[k] < 2
        x[k] = 0;
    elseif t[k] < 5
        x[k] = -8 + 4*t[k]
    elseif t[k] <7
        x[k] = 42 - 6*t[k]
    else
        x[k] = 0
    end
end
plot(t,x)
x=x*dt


for k = 1 : length(t)
    h[k] = 1.25*exp(-t[k])-1.25*exp(-5.0*t[k])
end

y = conv(x,h)
plot(t,y[1:length(t)])
```
