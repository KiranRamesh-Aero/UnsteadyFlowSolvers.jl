```julia
#Include the UNSflow libraries
#Pkg.clone("git://github.com/KiranUofG/UNSflow")
include("../src/UNSflow.jl")
using UNSflow
```

```julia
data = readdlm("naca0012_data_exp.txt"; skipstart = 1)
alpha = data[:,1]
CN = data[:,2];
CM = data[:,3];
```

```julia
f = fFromCN(alpha,CN);
```

```julia
alpha1, s1, s2, f_model = findStaticCoeff(alpha,f,"Sheng");
```

```julia
static_constants = cmstatic(CM, CN, f; par = 1.0)
```

```julia
nsteps = 500
alphamax = 20 
mat_case = zeros(nsteps, 10, alphamax+1)
for iter = 1:alphamax+1

            alphadef = ConstDef((iter-1)*pi/180)
            hdef = ConstDef(0.)
            udef = ConstDef(1.)#Freestream velocity is a constant. Set as constant = 1

            #The full kinematics are collected as:
            full_kinem = KinemDef(alphadef, hdef, udef)

            pvt = 0.25
            lespcrit = [5.;] #Note that the value is entered as an array. 

            # The argument "FlatPlate" provides zero camber. To use an airfoil, specify the coordinate file which should be
            #in the same folder. For eg. "sd7003.dat". 
            #The XFOIL convention must be used for the coordinate: TE->Upper surface->LE->Lower surface->TE
            surf = TwoDSurf("n0012.dat", pvt, full_kinem,lespcrit)

            curfield = TwoDFlowField()
            del = DelVortDef(1, 500, 10)
            #Specify simulation details - time step, and number of steps.
            #The default nondimensional timestep is 0.015. There is usually no need to change this. 

            #Now the code is run as :
            #The variables mat, surf and curfield are returned
            dtstar = 0.015
        
            mat, surf, curfield = ldvm_klb(surf, curfield,static_constants, nsteps, dtstar; delvort = del, alpha1 = alpha1, S1 = s1, S2 = s2, model="Sheng")

         
            mat_case[:,:,iter]=mat        
end
```

```julia
plot(mat_case[end,2,:]*180/pi,(mat_case[end,6,:]))
xlabel("Angle of attack in deg")
ylabel("Lift force coefficient CL")
legend()
```

```julia
plot(mat_case[end,2,:]*180/pi,(mat_case[end,6,:].*sin(mat_case[end,2,:])-mat_case[end,7,:])./cos(mat_case[end,2,:]))
xlabel("Angle of attack in deg")
ylabel("Chordwise coefficient CC")
legend()
```

```julia
plot(mat_case[end,2,:]*180/pi,mat_case[end,8,:])
xlabel("Angle of attack in deg")
ylabel("Moment coefficient CM")
legend()
```

```julia
plot(mat_case[end,2,:]*180/pi,mat_case[end,9,:], label="Model")
plot(alpha,f, label="Data from CN(a)")
xlabel("Angle of attack in deg")
ylabel("Separation point f, x/c")
legend()
```
