## This notebook compares static CL, CD and CM predictions between LDVM solver, LDVM_KLB solver, Xfoil and experimental data for SD7003 airfoil. 

#### To calculate forces coefficients, static constants $\alpha_1$, S1 and S2 must be obtained.
#### This can be done in various ways. First of all, separation point position may be determined basing on CN($\alpha$) characteristics from xfoil or from CF plots. Secondly, there are two different formulas for fitting function to choose from (Sheng and Original). Finally, as results from Xfoil significantly differ from experiments, obtained constants may be or may be not modified to fit to the experiment a little better.
#### Then, static constants for moment calculation need to be obtained.
#### Therefore 8 cases presented in the Table 1 are studied.

#### Table 1

| No | f is obtained from  |Model | Is $\alpha_1$ modified basing on experimental data? |
|:----:|:-------:|:-------------:|:--------------------------------:|
| 1  | CN( $\alpha$)  | Sheng et al. | no                             |
| 2  |CN( $\alpha$) | Original   | no                             |
| 3  |  CF plots | Sheng et al.   | no                            |
| 4  |  CF plots |Original    | no                            |
| 5 | CN($\alpha$) | Sheng et al.  | yes                             |
| 6 | CN($\alpha$) | Original   | yes                             |
| 7 | CF plots | Sheng et al.  | yes                            |
| 8 | CF plots | Original | yes                            |

#### The procedure is as follows:

A. Obtain f(alpha) values

B. Obtain static constants alpha1, s1, s2

C. Calculate static constants k0, k1, k2, m

```julia
#For help type ?functionname and refer to "KLB - static constants for f", "KLB - static constants for CM" notebooks.

#A.1 Obtain f from CN(alpha)

    #Read CN(alpha) data from file prepared in xfoil
    xfoil = Xfoil("sd7003_data.txt")

    #Calculate f basing on CN(alpha)
    f_CN = fFromCN(xfoil.alpha, xfoil.CN)

#A.2 Obtain f from CF plots 
    
    #Read prepared file
    data = readdlm("sd7003_data_cf.txt"; skipstart = 1)
    alpha_CF = data[:,1]
    f_CF = data[:,2]    

#B. Choose models for test cases
    model = ["Sheng", "Original",  "Sheng", "Original","Sheng", "Original",  "Sheng", "Original"]

    #Define empty arrays
    alpha1 = zeros(length(model))
    s1 = zeros(length(model))
    s2 = zeros(length(model))
    f_model_CN = zeros(Int(length(model)/2),length(xfoil.alpha))
    f_model_CF = zeros(Int(length(model)/2),length(alpha_CF))
    cm_st_const = zeros(length(model))

      #Obtain static constants for test cases 1-4
   for n = 1 : 2
           alpha1[n], s1[n],s2[n], f_model_CN[n,:] = findStaticCoeff(xfoil.alpha, f_CN, model[n])
    end
            
    for n = 3 : 4
            alpha1[n], s1[n],s2[n], f_model_CF[n-2,:] = findStaticCoeff(alpha_CF, f_CF, model[n])
    end
```

#### $\alpha_1$ can be approximated as the stall angle. As it may be seen in the image below, results from Xfoil significantly differ from experiments. $\alpha_1$ from experiments is ~10.5 deg, while Xfoil predicts $\alpha_1$ at ~14 deg. It is possible to change value of N (free transition criterion) in Xfoil from default 9 to 13, to provide better agreement with experimental results.
#### For this reason it is worth checking how the model would behave if the value of $\alpha_1$ was modified to fit to the experimental data. Values of S1 and S2 remain unchanged.


Source of experimental data: Selig, M.S., Lyon, C.A., Giguère, P., Ninham, C.N., and Guglielmo, J.J., Summary of Low-Speed Airfoil Data, Vol. 2, SoarTech Publications, Virginia Beach, VA, 1996, 252 pages. Wind tunnel data on 25 airfoils tested at Reynolds Numbers ranging from 40,000 to 400,000. Available at: http://m-selig.ae.illinois.edu/uiuc_lsat/Low-Speed-Airfoil-Data-V2.pdf

```julia
#B. continued Enter alpha1 from experiment (alpha1 can be approximated as the stall angle)
    alpha1_exp = 10.5
 
    for n = 5 : 8
            alpha1[n] = alpha1_exp
            s1[n]=s1[n-4]
            s2[n]=s2[n-4]
    end   
```

```julia
#C. Calculate moment cofficients: k0, k1, k2, m
    #Obtain f(alpha) for cases 5-8

    for n = 5 : 6
        f_model_CN[n-2,:] = fFromConst(xfoil.alpha, alpha1[n], s1[n], s2[n], model[n])
    end
            
cm_st_const = zeros(length(model),4)
cm_st_const[1,:] = cmstatic(xfoil.CM, xfoil.CN, f_model_CN[1,:])
cm_st_const[2,:] = cmstatic(xfoil.CM, xfoil.CN, f_model_CN[2,:])
cm_st_const[3,:] = cm_st_const[1,:]
cm_st_const[4,:] = cm_st_const[2,:]
cm_st_const[5,:] = cmstatic(xfoil.CM, xfoil.CN, f_model_CN[3,:])
cm_st_const[6,:] = cmstatic(xfoil.CM, xfoil.CN, f_model_CN[4,:])
cm_st_const[7,:] = cm_st_const[5,:]
cm_st_const[8,:] = cm_st_const[6,:];
```

### Convergence has been studied. Starting at ~600th iteration, high-amplitude peaks regularly appear. To limit computation time to minimum (as there were many cases to be studied) 500 iterations are calculated and a median of CL, CD and CM values is used.

### Test cases are now being run

```julia
    nsteps = 500
    cases = 1:8 #Choose test cases to be run
    alphamax = 20 

    mat_case = zeros(nsteps, 10, alphamax+1, length(cases))

for k = 1 : length(cases)

        for iter = 1:alphamax+1

            alphadef = ConstDef((iter-1)*pi/180)
            hdef = ConstDef(0.)
            udef = ConstDef(1.)#Freestream velocity is a constant. Set as constant = 1

            #The full kinematics are collected as:
            full_kinem = KinemDef(alphadef, hdef, udef)

            pvt = 0.25
            lespcrit = [5.;] #Note that the value is entered as an array. 

            # The argument "FlatPlate" provides zero camber. To use an airfoil, specify the coordinate file which should be
            #in the same folder. For eg. "sd7003.dat". 
            #The XFOIL convention must be used for the coordinate: TE->Upper surface->LE->Lower surface->TE
            surf = TwoDSurf("sd7003_fine.dat", pvt, full_kinem,lespcrit)

            curfield = TwoDFlowField()
            del = DelVortDef(1, 500, 10)
            #Specify simulation details - time step, and number of steps.
            #The default nondimensional timestep is 0.015. There is usually no need to change this. 

            #Now the code is run as :
            #The variables mat, surf and curfield are returned
            dtstar = 0.015
        
            mat, surf, curfield = ldvm_klb(surf, curfield,cm_st_const[k,:], nsteps, dtstar; delvort = del, alpha1 = alpha1[k], S1 = s1[k], S2 = s2[k], model=model[k])

         
            mat_case[:,:,iter, k]=mat        
        end
end 
```

### LDVM without KLB

```julia
nsteps = 500
    alphamax = 20 

    mat_ldvm = zeros(nsteps, 8, alphamax+1)

        for iter = 1:alphamax+1

            alphadef = ConstDef((iter-1)*pi/180)
            hdef = ConstDef(0.)
            udef = ConstDef(1.)#Freestream velocity is a constant. Set as constant = 1

            #The full kinematics are collected as:
            full_kinem = KinemDef(alphadef, hdef, udef)

            pvt = 0.25
            lespcrit = [5.;] #Note that the value is entered as an array. 

            # The argument "FlatPlate" provides zero camber. To use an airfoil, specify the coordinate file which should be
            #in the same folder. For eg. "sd7003.dat". 
            #The XFOIL convention must be used for the coordinate: TE->Upper surface->LE->Lower surface->TE
            surf = TwoDSurf("sd7003_fine.dat", pvt, full_kinem,lespcrit)

            curfield = TwoDFlowField()
            del = DelVortDef(1, 500, 10)
            #Specify simulation details - time step, and number of steps.
            #The default nondimensional timestep is 0.015. There is usually no need to change this. 

            #Now the code is run as :
            #The variables mat, surf and curfield are returned
            dtstar = 0.015
        
            mat, surf, curfield = ldvm(surf, curfield, nsteps, dtstar; delvort = del)
         
            mat_ldvm[:,:,iter]=mat        
        end
```
