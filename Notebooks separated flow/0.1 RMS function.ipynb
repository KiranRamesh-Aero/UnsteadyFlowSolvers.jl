Two sets of points are given, (x_1,y_1) and (x_2,y_2). Distribution of x_1 and x_2 is uneven, also min and max values of x_1 and x_2 may be different.

Objective is to calculate RMS error for a common part of x_1 and x_2. 


    Steps: a) Interpolate both functions. b) Let x_rms be common part of x_1 and x_2. c) Calculate RMS basing on x and interpolated functions.

```julia
x_1 = [0., 1., 3., 5., 7., 9., 10.]
x_2 = [0., 2. ,4.7, 6. ,7.1 ,8. ,9. ,10., 11., 12.];
y_1 = [0., 1., 3., 5., 7., 9., 10.]
y_2 = [0., 2.5,5.2, 6.8, 7.5, 8.8, 9.1, 10.3, 11.2, 12.5]
plot(x_1,y_1,"o",label="1")
plot(x_2,y_2,"x",label="2")
legend()
```

```julia
knots_1 = (x_1,) #syntax required by interpolate function
knots_2 = (x_2,)
itp_1 = interpolate(knots_1, y_1, Gridded(Linear())) 
itp_2 = interpolate(knots_2, y_2, Gridded(Linear()))

x_max_rms = min(x_1[end],x_2[end])
x_min_rms = max(x_1[1],x_2[1])
x_rms = linspace(x_min_rms, x_max_rms, 100)
diff = itp_2[x_rms].-itp_1[x_rms]

plot(x_rms,itp_1[x_rms],"b",label="1 itp")
plot(x_rms,itp_2[x_rms],"r",label="2 itp")
plot(x_1,y_1,"bo",label="1")
plot(x_2,y_2,"rx",label="2")
plot(x_rms,diff, label="diff")
legend()
```

```julia
s = 0.0
for a in diff
    s +=a*a
end

out = sqrt(s/length(diff))
```
